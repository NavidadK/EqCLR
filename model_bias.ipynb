```python
class RandomRightAngleRotation:
    """Randomly rotate PIL image by 90, 180, or 270 degrees."""
    def __call__(self, x):
        angle = int(torch.randint(1, 4, ()).item()) * 90
        return x.rotate(angle)
```

```python
perf_res = {}
perf_res_mean = {}

knns = []
for seed in range(3):
    file_path = f'results_rs/model_eval/5001_seed{seed}_resnet18_pathmnist_no_train_eval.pkl'
    
    with open(file_path, 'rb') as f:
        eval_eq = pickle.load(f)
    
    perf_res[f'ResNet18_{seed}'] = eval_eq
    
    knns.append(eval_eq['euclidean'][10])
perf_res_mean[f'ResNet']  = {}
perf_res_mean[f'ResNet']['KNN_mean'] = np.mean(np.array(knns))
perf_res_mean[f'ResNet']['KNN_std'] = np.std(np.array(knns))

perf_res_mean_df = pd.DataFrame(perf_res_mean).T
display(perf_res_mean_df)
```

```python
perf_eq = {}
perf_eq_mean = {}

for N in range(1, 18):
    knns = []
    for seed in range(5):
        file_path = f'results_rs/model_eval/4001_seed{seed}_no_training_path_mnist-eqCLR_resnet18_N{N}_eval.pkl'
        
        with open(file_path, 'rb') as f:
            eval_eq = pickle.load(f)
        
        perf_eq[f'eqResNet18_N{N}_{seed}'] = eval_eq
        
        knns.append(eval_eq['euclidean'][10])

    perf_eq_mean[f'eqResNet18_N{N}']  = {}
    perf_eq_mean[f'eqResNet18_N{N}']['KNN_mean'] = np.mean(np.array(knns))
    perf_eq_mean[f'eqResNet18_N{N}']['KNN_std'] = np.std(np.array(knns))

perf_eq_mean_df = pd.DataFrame(perf_eq_mean).T
display(perf_eq_mean_df)
```

```python
knns_best = [perf_eq_mean_df['KNN_mean'][f'eqResNet18_N{N}'] for N in [4, 8, 12, 16]]
print(np.mean(knns_best))
knns_even = [perf_eq_mean_df['KNN_mean'][f'eqResNet18_N{N}'] for N in range(1, 18) if N % 2 == 0]
knns_odd = [perf_eq_mean_df['KNN_mean'][f'eqResNet18_N{N}'] for N in range(1, 18) if N % 2 == 1]
print(f'Mean KNN for even N: {np.mean(knns_even)}')
print(f'Mean KNN for odd N: {np.mean(knns_odd)}')
```

```python
fig, ax = plt.subplots(figsize=(12, 5))

fontsize = 14
ax.errorbar(perf_res_mean_df.index, perf_res_mean_df['KNN_mean'], yerr=perf_res_mean_df['KNN_std'], fmt='--o', capsize=5, color='blue')

ax.errorbar([f'{n}' for n in range(1, 18)], perf_eq_mean_df['KNN_mean'], yerr=perf_eq_mean_df['KNN_std'], fmt='--o', capsize=5, color='blue', clip_on=False)

knns_even = [perf_eq_mean_df['KNN_mean'][f'eqResNet18_N{N}'] for N in range(1, 18) if N % 2 == 0]
knns_odd = [perf_eq_mean_df['KNN_mean'][f'eqResNet18_N{N}'] for N in range(1, 18) if N % 2 == 1]

ax.axhline(np.mean(knns_even), xmin=0.02, xmax=0.96, linestyle='--', linewidth=1.0, alpha=0.75, color='red', zorder=1, label='Mean even N')
ax.axhline(np.mean(knns_odd), xmin=0.02, xmax=0.96, linestyle='--', linewidth=1.0, alpha=0.75, color='green', zorder=1, label='Mean odd N')
ax.text(18.5, np.mean(knns_even) - 0.002, 'Mean even N', color='red', fontsize=fontsize, ha='center')
ax.text(18.5, np.mean(knns_odd) - 0.002, 'Mean odd N', color='green', fontsize=fontsize, ha='center')

ax.axvline(0.5, linestyle='--', linewidth=0.75, alpha=0.75, color='gray', zorder=1)

ax.set_ylim(0.44, 0.58)
ax.set_ylabel('KNN Accuracy', fontsize=fontsize)
ax.tick_params(axis='both', labelsize=fontsize)

ax.spines['bottom'].set_position(('outward', 10))
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
```
